```python
import requests
import json
from dotenv import load_dotenv
import os
import pandas as pd
load_dotenv()
```

```python
vic = {'type': 'Point', 'coordinates': [144.9868546, -37.8026587]}
```

```python
tok1 = os.getenv("tok1")
tok2 = os.getenv("tok2")
```

```python
#url = "https://favqs.com/api/session"
enlace = 'https://api.foursquare.com/v2/venues/explore'
starbucks_vic = "f55ed460-ee94-4800-87db-fc78ec7d13db"
```

```python
parametros = { "client_id" : tok1,
          "client_secret" : tok2,
          "v" : "20180323",
          "ll": f"{vic.get('coordinates')[1]},{vic.get('coordinates')[0]}",
          "query": f"{starbucks_vic}",
          "limit" : 10
}
```

```python
resp = requests.get (url = enlace, params = parametros)
data = json.loads(resp.text)
```
